```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("cereal.csv")
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
(df==-1).sum()
```

```python
df=df.replace(-1,np.NaN)
```

```python
for column in ['carbo','sugars','potass']:
    df[column]=df[column].fillna(df[column].mean())
```

```python
(df==-1).sum()
```

```python
{column:list(df[column].unique()) for column in ['mfr','type']}
```

```python
df['type']=df['type'].replace({'C':0,'H':1})
```

```python
df['type'].unique()
```

```python
dummies=pd.get_dummies(df['mfr'])
```

```python
df=pd.concat([df,dummies],axis=1)
```

```python
df=df.drop('mfr',axis=1)
```

```python
df
```

```python
for column in df.columns[-7:]:
    df[column]=df[column].astype(int)
```

```python
df
```

```python
y=df.loc[:,'rating']
x=df.drop('rating',axis=1)
```

```python
x=x.drop('name',axis=1)
x
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model=LinearRegression()
model_lr=Ridge(alpha=1.0)
model_lasso=Lasso(alpha=1.0)
```

```python
for model in [model,model_lr,model_lasso]:
    model.fit(x_train,y_train)
    print(model,model.score(x_test,y_test))
```
